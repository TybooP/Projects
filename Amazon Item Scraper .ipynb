```python
# import libraries

from bs4 import BeautifulSoup
import pandas as pd
import requests
import time
import datetime
import csv

import smtplib
```

```python
# connect to the Website

URL = 'https://www.amazon.co.uk/Mitre-5BB1118B78-Impel-Training-Football/dp/B09B27BWWW/ref=sxin_15_ac_d_rm?ac_md=0-0-Zm9vdGJhbGw%3D-ac_d_rm_rm_rm&content-id=amzn1.sym.eae3237b-3bc2-4b6f-9b26-f3c2840ee3a5%3Aamzn1.sym.eae3237b-3bc2-4b6f-9b26-f3c2840ee3a5&cv_ct_cx=football&keywords=football&pd_rd_i=B09B27BWWW&pd_rd_r=3b3823a1-f128-41d2-b4f5-aeac735c8484&pd_rd_w=AYJva&pd_rd_wg=jVWwg&pf_rd_p=eae3237b-3bc2-4b6f-9b26-f3c2840ee3a5&pf_rd_r=2NP4FH6M9Q7ZJHHRXJ7H&psc=1&qid=1657309002&sprefix=foot%2Caps%2C112&sr=1-1-e2b79f78-a3a0-4cac-b70d-2a5e4ae8e724'

headers = { "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/103.0.0.0 Safari/537.36", }

page = requests.get(URL, headers = headers)

soup1 = BeautifulSoup(page.content, "html.parser")
soup2 = BeautifulSoup(soup1.prettify(), "html.parser")

title = soup2.find(id="productTitle").get_text()
price = soup2.find(id="corePriceDisplay_desktop_feature_div",).get_text()

price_elements = []
for n in price:
    if n == " ":
        continue
    elif n == "\n":
        continue
    else:
        price_elements += n
        
correct_price = ''.join(price_elements)


rating = soup2.find(id="acrPopover").get_text()
number_of_ratings = soup2.find(id="acrCustomerReviewText").get_text()
```

```python
# a bit of formating

title = title.strip()


correct_price = correct_price.strip()[:6]
final_price = correct_price.strip()[1:]


rating = rating.strip()
number_of_ratings = number_of_ratings.strip()


print(title)
print(final_price)
print(rating)
print(number_of_ratings)
```

```python
today = datetime.date.today()

print(today)
```

```python
# putting the data into a csv
#ONLY RUN IT ONE TIME, OTHERWISE YOU DELETE YOUR ALREADY EXISTING DATA

header = ['Date','Title', 'Price', 'Rating', 'Number of ratings']
data = [today, title, final_price, rating, number_of_ratings]

with open('AmazonFootballWebScraperDataset.csv', 'w', newline = '', encoding = 'UTF8') as f:
    writer = csv.writer(f)
    writer.writerow(header)
    writer.writerow(data)
    
```
